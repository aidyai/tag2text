```python
def save_image_file_names_to_csv(folder_path, output_csv_file):
    """
    Save the file names of image files in the specified folder to a CSV file.

    Parameters:
        folder_path (str): Path to the folder containing image files.
        output_csv_file (str): Output CSV file name.
    """
    # Check if the folder path exists
    if not os.path.exists(folder_path):
        print(f"The specified folder '{folder_path}' does not exist.")
        return

    # Get the list of files in the folder
    image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    # Check if there are any image files in the folder
    if not image_files:
        print(f"No image files found in the folder '{folder_path}'.")
        return

    # Create and write to the CSV file
    with open(output_csv_file, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)

        # Write the header
        csv_writer.writerow(['File Name'])

        # Write the file names
        for file_name in image_files:
            csv_writer.writerow([file_name])

    print(f"File names have been saved to '{output_csv_file}'.")
```

```python
save_image_file_names_to_csv(folder_path, csv_file)
```

```python
%%writefile app.py

import pandas as pd
import streamlit as st
import json
import streamlit.components.v1 as components
import random


st.markdown("<h1 style='text-align: center; color: blue;'>data-annotator</h4>", unsafe_allow_html=True)

# Update the path to your new CSV file
#memes_path = "'/content/drive/MyDrive/chromav2.csv'"
#memes_folder = "/path/to/your/memes/folder/"
csv_path = "/content/drive/MyDrive/chromav2.csv"
folder_path =  "/content/drive/MyDrive/CHROMA/"


def load_data():
    return pd.read_csv(csv_path)

# Read the data
data = load_data()
data['image_caption'] = data['image_caption'].astype(str)

st.write("")
if "page" not in st.session_state:
    temp = data.loc[data['image_caption'] == 'nan']
    random_index = temp.sample(1).index[0]
    st.session_state.page = random_index

temp = data.loc[data['image_caption'] == 'nan']
remaining_count = temp.shape[0]
st.session_state.remaining = remaining_count

def next_page():
    temp = data.loc[data['image_caption'] == 'nan']
    remaining_count = temp.shape[0]
    random_index = temp.sample(1).index[0]
    st.session_state.page = random_index

def save():
    row_index = st.session_state.page
    data.at[row_index, 'image_caption'] = caption
    data.at[row_index, 'image_tag'] = modified_name
    data.to_csv(csv_path, index=False)

prev, middle, next = st.columns([6, 4, 6])

prev.write(f"Current index: {st.session_state.page}")
middle.button("Next", on_click=next_page)
next.write(f"Remaining {st.session_state.remaining} of {data.shape[0]}")

row = data.iloc[st.session_state.page]

name = row['image_caption']
image_path = row['File Name']  # Assuming the File_Name column contains the paths to images
temp_caption = row['image_tag']

st.write(name)

modified_name = st.text_input('image_tag', name)
caption = st.text_input('image_caption', temp_caption)

st.button("Save", on_click=save)
impath = folder_path + image_path
st.image(impath, width=300)
```
